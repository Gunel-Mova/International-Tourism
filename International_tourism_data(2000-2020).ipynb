```python
# Dropping 2020 column from original data frames
arrivals = arrivals.drop(['2020'], axis = 1)
inbound_expenditures = inbound_expenditures.drop(['2020'], axis = 1)
departures = departures.drop(['2020'], axis = 1)
outbound_expenditures = outbound_expenditures.drop(['2020'], axis = 1)
```

```python
# Interpolating the NaNs linearly
arrivals_mis = arrivals.iloc[:,4:24]
```

```python
# Selecting all rows with NaN under an entire DataFrame
arrivals_mis[arrivals_mis.isnull().all(axis=1)]
```

```python
arrivals_mis.dropna(how='all', inplace = True)
```

```python
arrivals_countries = arrivals.iloc[:, :4]
```

```python
arrivals_c = pd.concat([arrivals_countries,arrivals_mis], axis=1, join='inner')
```

```python
# Comparing missing values after interpolating. As you see, some rows still keep NaN values. 
arrivals_c.isna().sum(), arrivals.isna().sum()
```

```python
# Interpolating the NaNs linearly for other data frames
inbound_expenditures_mis = inbound_expenditures.iloc[:,4:24]
```

```python
inbound_expenditures_mis.interpolate(method='linear', axis =1, inplace=True)
```

```python
# Selecting all rows with NaN under an entire DataFrame
inbound_expenditures_mis[inbound_expenditures_mis.isnull().all(axis=1)]
inbound_expenditures_mis.dropna(how='all', inplace = True)
```

```python
i_expenditures = inbound_expenditures.iloc[:, :4]
```

```python
departures_mis = departures.iloc[:,4:24]
departures_mis.interpolate(method='linear', axis =1, inplace=True)
departures_countries = departures.iloc[:, :4]
departures_mis[departures_mis.isnull().all(axis=1)]
departures_mis.dropna(how='all', inplace = True)
departures_c = pd.concat([departures_countries,departures_mis], axis=1, join='inner')
```

```python
departures_c.isna().sum(), departures.isna().sum()
```

```python
outbound_mis = outbound_expenditures.iloc[:,4:24]
outbound_mis.interpolate(method='linear', axis =1, inplace=True)
outbound_mis[outbound_mis.isnull().all(axis=1)]
outbound_mis.dropna(how='all', inplace = True)
outbound_countries = outbound_expenditures.iloc[:, :4]
outbound_expenditures_c = pd.concat([outbound_countries,outbound_mis], axis=1, join='inner')
```

```python
outbound_expenditures_c.isna().sum(), outbound_expenditures.isna().sum()
```

There are some inconsistencies in Country Name column in all data frames. It doesn't consist of only countries, keeps regions and other categories. Let's check and create a subset of only countries.

```python
countries = arrivals_c["Country Name"].unique()
countries.sort()
countries
```

We have to delete all characters after separator ',' because it is not matching with iso 3166.

```python
arrivals_c["Country Name"] = arrivals_c["Country Name"].str.split(',').str[0]
```

```python
arrivals_c.replace("Korea","South Korea", inplace = True)
arrivals_c.head(1)
```

```python
from dataprep.clean import validate_country
```

```python
country = validate_country(arrivals_c["Country Name"])
country
```

```python
arrivals_countries = arrivals_c[validate_country(arrivals_c["Country Name"])]
```

```python
arrivals_countries.reset_index(drop = True, inplace = True)
```

```python
print(list(arrivals_countries["Country Name"]))
```

```python
inbound_expenditures_c["Country Name"] = inbound_expenditures_c["Country Name"].str.split(',').str[0]
inbound_expenditures_c.replace("Korea","South Korea", inplace = True)
inbound_expenditures_countries = inbound_expenditures_c[validate_country(inbound_expenditures_c["Country Name"])]
```

```python
inbound_expenditures_countries.reset_index(drop = True, inplace = True)
```

```python
arrivals_2020["Country Name"] = arrivals_2020["Country Name"].str.split(',').str[0]
arrivals_2020.replace("Korea","South Korea", inplace = True)
arrivals_2020 = arrivals_2020[validate_country(arrivals_2020["Country Name"])]
arrivals_2020.reset_index(drop = True, inplace = True)
```

```python
inboundexp_2020["Country Name"] = inboundexp_2020["Country Name"].str.split(',').str[0]
inboundexp_2020.replace("Korea","South Korea", inplace = True)
inboundexp_2020 = inboundexp_2020[validate_country(inboundexp_2020["Country Name"])]
inboundexp_2020.reset_index(drop = True, inplace = True)
```

```python
departures_c["Country Name"] = departures_c["Country Name"].str.split(',').str[0]
departures_c.replace("Korea","South Korea", inplace = True)
departures_countries = departures_c[validate_country(departures_c["Country Name"])]
departures_countries.reset_index(drop = True, inplace = True)
```

```python
outbound_expenditures_c["Country Name"] = outbound_expenditures_c["Country Name"].str.split(',').str[0]
outbound_expenditures_c.replace("Korea","South Korea", inplace = True)
outbound_expenditures_countries = outbound_expenditures_c[validate_country(outbound_expenditures_c["Country Name"])]
outbound_expenditures_countries.reset_index(drop = True, inplace = True)
```

```python
# Combining data frames on common columns
arrivals_inbound_exp = pd.merge(arrivals_countries, inbound_expenditures_countries, how = "inner", on = ["Country Name","Country Code"])
arrivals_inbound_exp.head()
```

```python
departures_outbound_exp = pd.merge(departures_countries,outbound_expenditures_countries, how = "inner", on = ["Country Name","Country Code"])
departures_outbound_exp.head()
```

```python
# Finding the top 10 countries that attract tourists
sorted_arrivals = arrivals_countries.sort_values(by = ["2017","2018","2019"], ascending = False).head(10)
```

```python
plt.plot(sorted_arrivals['Country Name'],sorted_arrivals['2017'],label='2017', color='green', linewidth=1)
plt.plot(sorted_arrivals['Country Name'],sorted_arrivals['2018'],label='2018', color='red',linestyle='dotted',linewidth=1)
plt.plot(sorted_arrivals['Country Name'],sorted_arrivals['2019'],label='2019', color='blue', linestyle='dashed', linewidth=1)
plt.legend()
plt.ylabel('Arrivals', fontsize=14)
plt.title('Arrivals of top countries', fontsize=16)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks(rotation=60)
plt.show()
```

```python
# Is there any positive correlation between tourists' number and inbound expenditures? 
```

```python
cor = sns.regplot(x="2017_x", y="2017_y", data=arrivals_inbound_exp, ci = None)
cor.set_title("Tourists vs. Expenditures")
cor.set_xlabel("Tourists")
cor.set_ylabel("Expenditures")
```

```python
correlation_matrix = arrivals_inbound_exp.corr()
round(correlation_matrix,4)
```

```python
sns.heatmap(correlation_matrix)
```

```python
# How has the tourism changed over years? How has Covid 19 impacted on Tourism?
```

```python
arrivals_countries_mean = arrivals_countries.mean()
arrivals_2020_mean = arrivals_2020.mean()
arrivals_countries_mean = pd.concat([arrivals_countries_mean, arrivals_2020_mean], axis=0, join = 'outer', sort = True)
arrivals_countries_mean.head()
```

```python
plt.plot(arrivals_countries_mean['2000':'2020'], color='blue', linewidth=2)
plt.ylabel('Tourists', fontsize=14)
plt.title(' The number of tourists over years', fontsize=16)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks(rotation=60)
plt.show()
```

```python
inbound_expenditures_countries_mean = inbound_expenditures_countries.mean()
inboundexp_2020_mean = inboundexp_2020.mean()
inbound_expenditures_mean = pd.concat([inbound_expenditures_countries_mean, inboundexp_2020_mean], axis=0, join = 'outer', sort = True)
```

```python
plt.plot(inbound_expenditures_mean['2000':'2020'], color='red', linewidth=2)
plt.ylabel('Expenditures', fontsize=14)
plt.title(' The amount of expenditure over years', fontsize=16)
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.xticks(rotation=60)
plt.show()
```
